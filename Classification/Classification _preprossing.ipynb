```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("gym_members_data_with_missing_values.csv")
```

```python
df.head(15)
```

```python
null_count = df.isnull().sum().sum()
print(f"Total missing values: {null_count}")
```

```python
numerical_features = ['Weight (kg)', 'Height (m)', 'Session_Duration (hours)']
categorical_features = ['Gender', 'Workout_Type']
```

```python
for col in numerical_features:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
for col in categorical_features:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df = pd.get_dummies(df, columns=categorical_features, drop_first=True)
```

```python
scaler = StandardScaler()
for col in numerical_features:
    df[col] = scaler.fit_transform(df[[col]])
```

```python
X = df.drop(columns=['Experience_Level'])  # Features
y = df['Experience_Level']  # Target
```

```python
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
df_resampled = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
print(df_resampled['Experience_Level'].value_counts())
```

```python
output_file_path_resampled = 'resampled_gym_members_data.csv'
df_resampled.to_csv(output_file_path_resampled, index=False)
```

```python
print(f"Processed data saved to: {output_file_path_resampled}")
```
